```python
import requests
import json
import pandas as pd
from key import api_key, domain
```

```python
def get_all__dep_users(domain, okta_api_token):
    error = None
    headers = {'Authorization': 'SSWS {}'.format(okta_api_token),
               'Accept': 'application/json',
               'Content-Type': 'application/json'}

    okta_url = "https://{}.okta.com".format(domain)
    url = '{0}/api/v1/users?filter=status eq "DEPROVISIONED"&limit=250'.format(okta_url)
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        users = response.json()
        links = response.links
        while 'next' in links:
            url = links['next']['url']
            response = requests.get(url, headers=headers)
            response.raise_for_status()
            next_users = response.json()
            users += next_users
            links = response.links
    except Exception as e:
        error = "get_all_users failed with exception {}".format(e)
        print(error)
    return (users)
```

```python
def get_all_users(domain, okta_api_token):
    error = None
    headers = {'Authorization': 'SSWS {}'.format(okta_api_token),
               'Accept': 'application/json',
               'Content-Type': 'application/json'}

    okta_url = "https://{}.okta.com".format(domain)
    url = '{0}/api/v1/users?imit=5'.format(okta_url)
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        users = response.json()
        links = response.links
        while 'next' in links:
            url = links['next']['url']
            response = requests.get(url, headers=headers)
            response.raise_for_status()
            next_users = response.json()
            users += next_users
            links = response.links
    except Exception as e:
        error = "get_all_users failed with exception {}".format(e)
        print(error)
    return (users)
```

```python
def delete_user(domain, okta_api_token, user):
    error = None
    headers = {'Authorization': 'SSWS {}'.format(okta_api_token),
               'Accept': 'application/json',
               'Content-Type': 'application/json'}
    
    okta_url = "https://{}.okta.com".format(domain)
    url = '{0}/api/v1/users/{1}'.format(okta_url,user)
    try:
        response = requests.delete(url, headers=headers)
        response.raise_for_status()

    except Exception as e:
        error = "get_all_users failed with exception {}".format(e)
        print(error)
    return (response.raise_for_status())
```
